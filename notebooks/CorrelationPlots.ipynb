```python
import pandas as pd
import seaborn as sns
import pickle
# from pandas.plotting import scatter_matrix
import plotly.plotly as py
import plotly
import plotly.figure_factory as ff
```

```python
alberta = pd.read_pickle("../data/alberta.pkl")
florida = pd.read_pickle("../data/florida.pkl")
df = pd.concat([alberta, florida], sort=False).reset_index(drop=True)
#df.head()
```

```python
sub_df = df[['Secchi Depth (m)', 'Microcystin (ug/L)',
             'Total Phosphorus (ug/L)', 'Total Nitrogen (ug/L)',
             'Altitude']].rename(columns = {'Secchi Depth (m)' : 'Sechhi Depth',
                         'Microcystin (ug/L)' : '[MC]',
                         'Total Phosphorus (ug/L)' : '[P]',
                         'Total Nitrogen (ug/L)' : '[N]'})

fig = ff.create_scatterplotmatrix(sub_df, height=800, width=800)
#plotly.offline.iplot(fig, validate=False)
# print(fig['data'][0]['yaxis'])
```

```python
sub_df1 = sub_df[sub_df['[MC]'] <= 4]
fig = ff.create_scatterplotmatrix(sub_df1, height=800, width=800)
plotly.offline.iplot(fig, validate=False)
```

```python
sub_df2 = sub_df[(sub_df['[MC]'] > 4) & (sub_df['[MC]'] <= 20)]
fig = ff.create_scatterplotmatrix(sub_df2, height=800, width=800)
plotly.offline.iplot(fig, validate=False
```

```python
sub_df3 = sub_df[sub_df['[MC]'] > 20]
fig = ff.create_scatterplotmatrix(sub_df3, height=800, width=800)
plotly.offline.iplot(fig, validate=False)
```
